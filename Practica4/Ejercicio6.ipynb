# 6. Usar la iteración de Punto Fijo para mostrar que la función $g(x) = π + 0.5 sin(x/2)$ tiene un punto fijo en el intervalo [0, 2π] con una tolerancia de 10^-2. (Programar el ejemplo en específico con una gráfica de la función).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
  return np.pi + 0.5 * np.sin(x/2)

tolerancia = 0.01
iteraciones = [1.0]
errores = []
```

Se define la iteracion de Punto Fijo en un ciclo while y se imprimen los valores $x_n$ que obtenemos

```python
#Iteracion Punto Fijo
while True:
    x_nuevo = g(iteraciones[-1])
    error = abs(x_nuevo - iteraciones[-1])
    iteraciones.append(x_nuevo)
    errores.append(error)
    if error < tolerancia:
        break

#Impresion de los valores
for i in range(0, len(iteraciones)):
    print(f"x{i} = {iteraciones[i]:.5f}")

x_final = iteraciones[-1]
print(f"\nPunto fijo aproximado: x = {x_final:.5f} con tolerancia 10^-2")
```

Se grafica la función en el rango establecido por el problema

```python
x_vals = np.linspace(0, 2 * np.pi, 400)
g_vals = g(x_vals)
identity_vals = x_vals

plt.plot(x_vals, g_vals, label="g(x) = π + 0.5 sin(x/2)", linewidth=2)
plt.plot(x_vals, identity_vals, label="y = x", linewidth=2)
plt.axvline(x=0, color='gray', linestyle='--', linewidth=1, label="x = 0")
plt.axvline(x=2 * np.pi, color='gray', linestyle='--', linewidth=1, label="x = 2π")
plt.scatter(x_final, x_final, color='red', label=f"Punto fijo = {x_final:.5f}")
plt.title("Iteración de Punto Fijo g(x) y la recta y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
