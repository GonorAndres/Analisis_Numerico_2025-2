### Programa el producto punto de dos vectores en Rn, por bloques. Note que para que esto tenga relevacia computacional, la n debe ser mayor que 50.

```python
import numpy as np
import timeit

# Producto punto de dos vectores
def pp_bloques(M,N):
  if M.shape[0] != N.shape[0]:
    print("No se puede realizar la operaci√≥n")
  else:
    #obtenemos los puntos medios

    a1 = M.shape[0] // 2

    # Dividimos las vectores en bloques
    M1 = M[a1:,:] # izquierda
    M2 = M[:a1,:] # derecha

    N1 = N[a1:,:] # izquierda
    N2 = N[:a1,:] # derecha

    C1, C2 = 0, 0
    # Calculamos los bloques por cada bloque dividido
    for i in range(M1.shape[0]):
      C1 += M1[i,0] * N1[i,0]

    for i in range(M2.shape[0]):
      C2 += M2[i,0] * N2[i,0]

    C = C1 + C2
    return float(C)
```

```python
#Creamos dos vectores de la misma dimension
v1 = np.random.rand(60,1)
v2 = np.random.rand(60,1)

pp_bloques(v1,v2)
```
