### Programe la suma de matrices por bloques.

```python
import numpy as np
import timeit

# Suma de matrices por bloques
def summult_bloques(M,N):
  if M.shape != N.shape:
    print("No se puede sumar estas matrices")
  else:
    #obtenemos los puntos medios

    a1 = M.shape[0] // 2
    a2 = M.shape[1] // 2
    b1 = N.shape[0] // 2
    b2 = N.shape[1] // 2

    # Dividimos las matrices en bloques
    M11 = M[:a1, :a2] # superior izquierda
    M12 = M[:a1, a2:] # superior derecha
    M21 = M[a1:, :a2] # inferior izquierda
    M22 = M[a1:, a2:] # inferior derecha

    N11 = N[:b1, :b2] # superior izquierda
    N12 = N[:b1, b2:] # superior derecha
    N21 = N[b1:, :b2] # inferior izquierda
    N22 = N[b1:, b2:] # inferior derecha

    # Calculamos los bloques de la matriz C usando el proceso similiar a la multiplicacion de matrices
    C11 = M11 + N11
    C12 = M12 + N12
    C21 = M21 + N21
    C22 = M22 + N22

    # Combinamos los bloques para formar la matriz C completa
    C = np.block([[C11, C12], [C21, C22]])
    return C
```

```python
A = np.random.rand(80,80)
B = np.random.rand(80,80)

# Medir el tiempo de ejecución
time = timeit.timeit(lambda: A + B, number=1000)

print(f"Tiempo de ejecución: {time:.6f} segundos para 1000 ejecuciones")

time = timeit.timeit(lambda: summult_bloques(A,B), number=1000)

print(f"Tiempo de ejecución: {time:.6f} segundos para 1000 ejecuciones")
```
